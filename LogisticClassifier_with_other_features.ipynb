```python
import pandas as pd

sp500 = pd.read_csv(r"/home/lpcheung/Downloads/sp500.csv")
gold = pd.read_csv(r"/home/lpcheung/Downloads/gold-price.csv")
oil = pd.read_csv(r"/home/lpcheung/Downloads/oil-price.csv")
hsi = pd.read_csv(r"/home/lpcheung/Downloads/hsi.csv")
```

```python
sp500['Date'] = pd.to_datetime(sp500['Date'])
gold['Date'] = pd.to_datetime(gold['Date'])
oil['DATE'] = pd.to_datetime(oil['DATE'])
hsi['Date'] = pd.to_datetime(hsi['Date'])

gold.index = gold['Date']
oil.index = oil['DATE']
hsi.index = hsi['Date']
sp500.index = sp500['Date']
```

```python
hsi = hsi.merge(sp500, left_on = 'Date', right_on = 'Date', how = 'inner')
```

```python
hsi = hsi.merge(gold, left_on = 'Date', right_on = "Date", how = "inner")
```

```python
hsi = hsi.merge(oil, left_on = 'Date', right_on = "DATE", how = "inner")
```

```python
x_train = []
y_train = []
array = hsi.values
for i in range(len(hsi.values)-3):
    temp = []
    try:
        for j in range(3):

            temp.append(array[i+j+1][2])
            temp.append(array[i+j+1][11])
            temp.append(array[i+j+1][17])
            temp.append(float(array[i+j+1][19]))
        x_train.append(temp)
        y_train.append(array[i][2]>array[i+1][2])
    except:
        pass
```

```python
from sklearn import linear_model
predictor = linear_model.LogisticRegression()

predictor.fit(x_train, y_train)
```

```python
predictor.score(x_train, y_train)
```
